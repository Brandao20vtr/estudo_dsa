# Reduce
Recebe 2 argumentos.
- Primeiro argumento é o nome de uma função.
- Segundo argumento é uma sequência.

Aplica a função passada como parâmetro aos elementos da sequência até que reste apenas um elemento.



**Ex:** Começa no valor 47 e soma mais 11 que é igual a 58, em seguida pega o 58 e soma com 42 que é 100 e por fim pega 100 e soma 13 que retorna 113.

```python
# Por mais que essa função Reduce seja buildin, ela deve ser importada
from functools import reduce
```

```python
# Criando uma lista
lista = [15, 65, 7, 11]

# Função para soma
def soma(a, b):
    x = a + b
    return x
```

```python
# Aplicando a função soma no conjunto de dados
reduce(soma, lista)
```

```python
# Criando outra lista
lst = [22, 9, 5, 30]

# Reduce com uma função anônima
reduce(lambda x,y: x+y, lst)
```

```python
# A função Reduce permite você pegar uma lista contendovários
# elementos e recuzí-lá em 1 elemento.
```

```python
# Atribuindo lambda a uma variável para encontrar o maior valor da lista
maxFind = lambda x,y: x if (x > y) else y
```

```python
reduce(maxFind, lst)
```
